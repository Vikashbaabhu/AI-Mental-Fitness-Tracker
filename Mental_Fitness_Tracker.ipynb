# 1. Take the data and create dataframe

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('mental-and-substance-use-as-share-of-disease.csv')
df2 = pd.read_csv('prevalence-by-mental-and-substance-use-disorder.csv')
```

```python
df1.head()
```

```python
df2.head()
```

# 2. Filtering the Data (Data Cleaning, Dropping)

```python
df1 = df1.fillna(df1.mean())
df2 = df2.fillna(df2.mean())
```

```python
df = pd.merge(df1,df2)
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.drop('Code',axis=1,inplace=True)
```

```python
df.set_axis(['Country','Year','Mental Fitness','Schizophrenia','Bipolar disorder','Eating disorder','Anxiety','Drug','Depressive disorder','Alcohol'],axis='columns',inplace=True)
```

# 4. Divide the data into input and output

```python
# Input - Country, Year, Schizophrenia, Bipolar disorder, Eating disorder, Anxiety, Drug, Depressive disorder, Alcohol
# Output - Mental Fitness
```

```python
# Converting the object data type to integer data type

from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
for i in df.columns:
  if df[i].dtype == 'object':
    df[i] = l.fit_transform(df[i])
```

```python
df.info()
```

```python
df.shape
```

```python
#Input

x = df.drop('Mental Fitness',axis=1)
x
```

```python
#Output

y = df['Mental Fitness']
y
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python

#After train_test_split ,

# Training varaibles get 80% data of x and y
# Testing variables get 20% data from x and y
```

```python
print('xtrain',xtrain.shape)
print('xtest',xtest.shape)
print('\nytrain',ytrain.shape)
print('ytest',ytest.shape)
```

#6. Running Regressor

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

#7.Fit the model

```python
lr.fit(xtrain,ytrain)
```

#8.Predict the output

```python
y_pred = lr.predict(xtest)
y_pred
```

```python
# ACTUAL OUTPUT VALUES
ytest
```

#9.Evaluating the Accuracy Score

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(ytest,y_pred)
rmse = (np.sqrt(mse))
r2 = r2_score(ytest,y_pred)
```

```python
print('Linear Regression model performance')
print('-----------------------------------')
print('MSE is',mse)
print('RMSE is',rmse)
print('R2 score is',r2)
```

# **Running other Regressors**

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain,ytrain)

y_pred = rf.predict(xtest)
mse = mean_squared_error(ytest,y_pred)
rmse = (np.sqrt(mse))
r2 = r2_score(ytest,y_pred)
```

```python
print('Random Forest Regressor model performance')
print('-----------------------------------------')
print('MSE is',mse)
print('RMSE is',rmse)
print('R2 score is',r2)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(xtrain, ytrain)
y_pred = dt.predict(xtest)
mse = mean_squared_error(ytest,y_pred)
rmse = (np.sqrt(mse))
r2 = r2_score(ytest,y_pred)
```

```python
print('Decision Tree Regressor model performance')
print('-----------------------------------------')
print('MSE is',mse)
print('RMSE is',rmse)
print('R2 score is',r2)
```

# K-Nearest Neighbors Regressor

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor()
knn.fit(xtrain, ytrain)
y_pred = knn.predict(xtest)
mse = mean_squared_error(ytest,y_pred)
rmse = (np.sqrt(mse))
r2 = r2_score(ytest,y_pred)
```

```python
print('K-Nearest Neighbors Regressor model performance')
print('-----------------------------------------')
print('MSE is',mse)
print('RMSE is',rmse)
print('R2 score is',r2)
```

# XGBoost Regressor

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()
xgb.fit(xtrain, ytrain)
y_pred = xgb.predict(xtest)
mse = mean_squared_error(ytest,y_pred)
rmse = (np.sqrt(mse))
r2 = r2_score(ytest,y_pred)
```

```python
print('XGBoost Regressor model performance')
print('-----------------------------------------')
print('MSE is',mse)
print('RMSE is',rmse)
print('R2 score is',r2)
```
